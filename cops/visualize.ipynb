```python
import matplotlib.pyplot as plt
import numpy as np

models = ['numpy', 'core', 'scikit']

mae = [48515.6537, 48515.6534, 48512.8409]
rmse = [68558.2029, 68558.2024, 67730.1064]
r2 = [0.6774, 0.6774, 0.6499]

x = np.arange(len(models))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(x - width, mae, width, label='MAE')
bar2 = ax.bar(x, rmse, width, label='RMSE')
bar3 = ax.bar(x + width, r2, width, label='R² Score')

ax.set_xlabel('Models')
ax.set_title('Regression Metrics Comparison Across Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Add value labels on top of bars
def add_labels(bars, fmt='{:.2f}'):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(fmt.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # offset label position
                    textcoords='offset points',
                    ha='center', va='bottom', fontsize=9)

add_labels(bar1, '{:.0f}')
add_labels(bar2, '{:.0f}')
add_labels(bar3, '{:.3f}')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['numpy', 'core', 'scikit']

mae = [48515.6530, 48515.6534, 48512.8409]
rmse = [68558.2018, 68558.2024, 67730.1064]
r2 = [0.6774, 0.6774, 0.6499]

x = np.arange(len(models))
width = 0.5

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# MAE plot
axs[0].bar(x, mae, width, color='skyblue')
axs[0].set_xticks(x)
axs[0].set_xticklabels(models)
axs[0].set_title('MAE Comparison')
axs[0].set_ylabel('MAE')
for i, v in enumerate(mae):
    axs[0].text(i, v + 1000, f'{v:.0f}', ha='center')

# RMSE plot
axs[1].bar(x, rmse, width, color='lightgreen')
axs[1].set_xticks(x)
axs[1].set_xticklabels(models)
axs[1].set_title('RMSE Comparison')
axs[1].set_ylabel('RMSE')
for i, v in enumerate(rmse):
    axs[1].text(i, v + 1000, f'{v:.0f}', ha='center')

# R2 plot
axs[2].bar(x, r2, width, color='salmon')
axs[2].set_xticks(x)
axs[2].set_xticklabels(models)
axs[2].set_title('R² Score Comparison')
axs[2].set_ylabel('R² Score')
for i, v in enumerate(r2):
    axs[2].text(i, v + 0.01, f'{v:.3f}', ha='center')

plt.tight_layout()
plt.show()
```
